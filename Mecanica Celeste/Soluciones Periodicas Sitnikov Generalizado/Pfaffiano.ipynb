# Computos de Pfaffiano

```python
from sympy import *
orden=6
a=symbols('a:orden(:orden)')
L=[]
for i in range(orden):
    L1=[]
    for j in range(orden):
        if j<i:
            L1+=[-a[i*orden+j]]
        elif i==j:
            L1+=[0]
        else:
            L1+=[a[i*orden+j]]
    L+=[L1]
M=Matrix(L)    
```

```python
from sympy import *
orden=6
Simb='a0_0'
for i in range(orden):
    for j in range(orden):
        if i+j>0:
            Simb+=',a'+str(i)+'_'+str(j)
Simb=symbols(Simb)
```

```python
L=[]
for i in range(orden):
    L1=[]
    for j in range(orden):
        if j<i:
            L1+=[-Simb[j*orden+i]]
        elif i==j:
            L1+=[0]
        else:
            L1+=[Simb[i*orden+j]]
    L+=[L1]
M=Matrix(L)
```

```python
init_printing()
M
```

```python
detA=M.det()
detA
```
