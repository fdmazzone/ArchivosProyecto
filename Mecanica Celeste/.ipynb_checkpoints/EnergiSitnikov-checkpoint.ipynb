```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
```

```python
x=np.arange(-10,10,.1)
v=np.arange(-5,5,.05)
X,V=np.meshgrid(x,v)
E=.5*V**2-sum([1/(r**2+X**2) for r in np.arange(1,10,1)])
```

```python
fig=plt.figure()
ax=Axes3D(fig)
#Z=np.minimum(Z,100*np.ones(np.shape(Z)))
#Z=np.maximum(Z,-1*np.ones(np.shape(Z)))
ax.plot_surface(X,V,E,rstride=1, cstride=1, cmap=cm.coolwarm,linewidth=0, antialiased=False)
#V=[k for k in [-20,0,20]]#np.arange(-20,20,1)]
#cp = plt.contour(X, Y, Z,V)
#plt.clabel(cp, inline=True,fontsize=10)
plt.xlabel('X')
plt.ylabel('V')
plt.show()
```

```python
fig=plt.figure()
H=np.concatenate((np.arange(-1,0,.1),np.arange(0,10,1)))
cp = plt.contour(X, V, E,H)
plt.clabel(cp, inline=True,fontsize=10)
plt.xlabel('X')
plt.ylabel('V')
plt.show()
```
