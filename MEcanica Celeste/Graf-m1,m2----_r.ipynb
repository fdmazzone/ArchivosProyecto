```python
from matplotlib.pyplot import *
from mpl_toolkits.mplot3d import *
from pylab import *
from numpy import *
```

```python
Cant_Cuerpos=3
def func(m):
    L=[m[0]+m[1],3*m[0]+2*m[1], 3*m[0]+m[1], -(m[1]+3*m[2]), -(2*m[1]+3*m[2]), -(m[1]+m[2])]
    sol=roots(L)
    for s in sol:
        if s.imag==0 and s.real>0:
            Solreal=s.real
    return  Solreal
```

```python
M1=arange(0.01,20,.1)
M2=arange(.01,20,.1)
Z=zeros([shape(M1)[0],shape(M2)[0]])
for  i in range(shape(M1)[0]):
    for j in range(shape(M2)[0]):
        Z[i,j]=func([1,M1[i],M2[j]])
Z=minimum(Z,1000*ones(shape(Z)))
X,Y=meshgrid(M1,M2)
fig=figure()
ax=Axes3D(fig)
ax.plot_surface(X,Y,Z.transpose())
xlabel('m1')
ylabel('m2')
show()
```

```python
fig=figure()
V=[k for k in arange(0,10,.1)]
cp = contour(X, Y, Z.transpose(),V)
clabel(cp, inline=True,fontsize=10)
title('Contour Plot')
xlabel('m1')
ylabel('m2')

show()
```

<h2> La funci√≥n $r=r(m_1,m_2)$ parece concava respecto a $m_2$ si fijamos $m_1$. Chequeemos que la derivada $\frac{\partial r}{\partial m_2}$ es decreciente </h2>

```python
h=.1
M1=arange(0.01,20,.1)
M2=arange(.01,20,.1)
Z=zeros([shape(M1)[0],shape(M2)[0]])
for  i in range(shape(M1)[0]):
    for j in range(shape(M2)[0]-1):
        Z[i,j]=(func([1,M1[i],M2[j+1]])-func([1,M1[i],M2[j]]))*h**(-1)
```

```python
M1,M2,Z=M1[:-1],M2[:-1],Z[:-1,:-1]
X,Y=meshgrid(M1,M2)
fig=figure()
ax=Axes3D(fig)
ax.plot_surface(X,Y,Z.transpose())
xlabel('m1')
ylabel('m2')
show()
```

```python
fig=figure()
V=[k**2 for k in arange(0,1,.01)]
cp = contour(X, Y, Z.transpose(),V)
clabel(cp, inline=True,fontsize=10)
title('Contour Plot')
xlabel('m1')
ylabel('m2')

show()
```
